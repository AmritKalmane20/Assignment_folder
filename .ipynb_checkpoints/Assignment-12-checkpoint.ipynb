```python
import matplotlib.pyplot as plt
import pandas as pd

data = [
    ['E001', 'M', 34, 123, 'NORMAL', 350],
    ['E002', 'F', 40, 114, 'OVERWEIGHT', 450],
    ['E003', 'F', 37, 135, 'OBESITY', 169],
    ['E004', 'M', 30, 139, 'UNDERWEIGHT', 189],
    ['E005', 'F', 44, 117, 'UNDERWEIGHT', 183]
            ]

cols = ['EPID', 'GENDER', 'AGE', 'SALES', 'BMI', 'INCOME']
df = pd.DataFrame(data, columns=cols)


df.plot.box()
plt.boxplot(df['INCOME'])
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate a sample dataset
np.random.seed(42)

groups = ['Group A', 'Group B', 'Group C', 'Group D']
male_ratio = np.random.uniform(0, 1, size=len(groups))
total_ratio = male_ratio + 1 
female_ratio = 1 - male_ratio/ total_ratio

data = pd.DataFrame({
    'Group': groups,
    'Male Ratio': male_ratio,
    'Female Ratio': female_ratio
})

# Save the dataset to a CSV file
data.to_csv('gender_ratio_dataset.csv', index=False)

# Plotting

# Plotting

# Pie Chart
plt.figure(figsize=(4, 4))
plt.pie(data['Male Ratio'], labels=data['Group'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral', 'lightgreen', 'lightpink'])
plt.title('Male Ratio in Different Groups')
plt.show()

# Column Chart for Male Ratio
plt.figure(figsize=(4, 4))
plt.bar(data['Group'], data['Male Ratio'], color='skyblue')
plt.xlabel('Groups')
plt.ylabel('Male Ratio')
plt.title('Male Ratio in Different Groups (Column Chart)')
plt.show()

# Histogram for Male Ratio
plt.figure(figsize=(4, 4))
plt.hist(data['Male Ratio'], bins=10, color='lightblue', edgecolor='black')
plt.xlabel('Male Ratio')
plt.ylabel('Frequency')
plt.title('Distribution of Male Ratio')
plt.show()

# Scatter Chart for Male Ratio
plt.figure(figsize=(4, 4))
plt.scatter(data['Group'], data['Male Ratio'], color='blue')
plt.xlabel('Groups')
plt.ylabel('Male Ratio')
plt.title('Male Ratio in Different Groups (Scatter Chart)')
plt.show()

# Boxplot for Male Ratio
plt.figure(figsize=(4, 4))
plt.boxplot(data['Male Ratio'])
plt.xticks([1], ['Male Ratio'])
plt.title('Boxplot of Male Ratio')
plt.show()

# Now let's repeat the charts for Female Ratio

# Pie Chart for Female Ratio
plt.figure(figsize=(4, 4))
plt.pie(data['Female Ratio'], labels=data['Group'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral', 'lightgreen', 'lightpink'])
plt.title('Female Ratio in Different Groups')
plt.show()

# Column Chart for Female Ratio
plt.figure(figsize=(4, 4))
plt.bar(data['Group'], data['Female Ratio'], color='lightcoral')
plt.xlabel('Groups')
plt.ylabel('Female Ratio')
plt.title('Female Ratio in Different Groups (Column Chart)')
plt.show()

# Histogram for Female Ratio
plt.figure(figsize=(4, 4))
plt.hist(data['Female Ratio'], bins=10, color='lightcoral', edgecolor='black')
plt.xlabel('Female Ratio')
plt.ylabel('Frequency')
plt.title('Distribution of Female Ratio')
plt.show()

# Scatter Chart for Female Ratio
plt.figure(figsize=(4, 4))
plt.scatter(data['Group'], data['Female Ratio'], color='lightcoral')
plt.xlabel('Groups')
plt.ylabel('Female Ratio')
plt.title('Female Ratio in Different Groups (Scatter Chart)')
plt.show()

# Boxplot for Female Ratio
plt.figure(figsize=(4, 4))
plt.boxplot(data['Female Ratio'])
plt.xticks([1], ['Female Ratio'])
plt.title('Boxplot of Female Ratio')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate a sample dataset
np.random.seed(42)
total_samples = 100
male_ratio = np.random.uniform(0, 1, total_samples)
female_ratio = 1 - male_ratio

# Create a DataFrame or use NumPy arrays directly depending on your data structure
# Let's assume you have other numeric columns in your dataset as well
# For simplicity, we'll only use male_ratio and female_ratio in this example
data = np.column_stack((male_ratio, female_ratio))

# Plotting

# Pie chart
labels = ['Male', 'Female']
plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
plt.pie(data.sum(axis=0), labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')

# Column chart
plt.subplot(2, 3, 2)
x = np.arange(len(labels))
width = 0.35
plt.bar(x, data.sum(axis=0), width, label='Total')
plt.xlabel('Gender')
plt.ylabel('Total Ratio')
plt.xticks(x, labels)
plt.title('Column Chart')

# Histogram
plt.subplot(2, 3, 3)
plt.hist(data, bins=10, label=labels)
plt.xlabel('Ratio')
plt.ylabel('Frequency')
plt.legend()
plt.title('Histogram')

# Scatter chart
plt.subplot(2, 3, 4)
plt.scatter(data[:, 0], data[:, 1])
plt.xlabel('Male Ratio')
plt.ylabel('Female Ratio')
plt.title('Scatter Chart')

# Boxplot
plt.subplot(2, 3, 5)
plt.boxplot(data, labels=labels)
plt.xlabel('Gender')
plt.ylabel('Ratio')
plt.title('Boxplot')

plt.tight_layout()
plt.show()
```
