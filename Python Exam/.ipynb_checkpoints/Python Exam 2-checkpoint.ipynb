```python
import numpy as np

# Assuming exam scores dataset as an array
exam_scores = np.array([78, 85, 90, 92, 88, 76, 80, 95, 82, 89])

# a) Calculate mean and median
mean_score = np.mean(exam_scores)
median_score = np.median(exam_scores)

# b) Compute standard deviation and variance
std_dev = np.std(exam_scores)
variance = np.var(exam_scores)

print(f"Mean: {mean_score}, Median: {median_score}")
print(f"Standard Deviation: {std_dev}, Variance: {variance}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
data=np.array([10,20,30,40,50,60,70,80,90,95])
plt.hist(data, bins=[0,25,50,75,100],edgecolor='black')
plt.xlabel('Score Bins')
plt.ylabel('Frequency')
plt.title('Histogram of Exam Scores')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
experience=np.array([2,5,8,3,6,4,7])
salary=np.array([50000,70000,80000,90000,60000,25000,65000])
#display scatter plot
plt.scatter(experience, salary, marker='*', color='purple')
plt.xlabel('Professional Experience')
plt.ylabel('Salary')
plt.title('Scatter Plot: Professional Experience vs Professional Salary')
plt.show()
```

```python
height=1.75
weight=70
bmi=weight/(height**2)

if bmi < 18.5:
    health_status = 'Underweight'
elif 18.5 <= bmi < 25:
    health_status = 'Normal Weight'
elif 25 <= bmi <30:
    health_status = 'Overweight'
else:  
    health_status = 'Obese'

print(f"BMI: {bmi}, Health Status: {health_status}")
```
